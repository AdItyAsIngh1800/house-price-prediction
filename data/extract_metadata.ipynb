```python
import pandas as pd
df = pd.read_csv('price_paid_records.csv')
# Take the first 5 rows and a random sample of 5
sample = pd.concat([df.head(5), df.sample(5)])

# Convert to markdown string
markdown_sample = sample.to_markdown()

# Save to a text file
with open('dataset_sample.md', 'w') as f:
    f.write(markdown_sample)

print("Sample saved to dataset_sample.md")
```
